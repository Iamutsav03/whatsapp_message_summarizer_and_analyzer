```python
import pandas as pd
import numpy as np
```

```python
with open('WhatsApp Chat with only kaam (btech IT unofficial).txt','r' ,encoding='utf-8') as f:
    data = f.read()
print(data)
```

```python
import re
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'

dates = re.findall(pattern, data)
messages = re.split(pattern, data)[1:]

print(messages)
print(dates)
```

```python
df = pd.DataFrame({'user_messages': messages, 'dates': dates})

df['dates'] = pd.to_datetime(
    df['dates'],
    format='%d/%m/%Y, %H:%M - '
)
```

```python
users = []
messages_clean = []

for message in df['user_messages']:
    split = message.split(': ', 1)
    
    if len(split) == 2:
        user, msg = split
        users.append(user)
        messages_clean.append(msg.strip())
    else:
        users.append('group_notification')
        messages_clean.append(message.strip())
```

```python
df['users'] = users
df['user_messages'] = messages_clean
```

```python
def clean_user(x):
    if re.match(r'\+?\d+', x):
        return f'User {x[-4:]}'   
    return x



df['user_messages'] = (
    df['user_messages']
    .astype(str)
    .str.replace('\n', ' ')
    .str.strip()
)

df['users'] = df['users'].astype(str).str.strip()

df['users'] = df['users'].apply(clean_user)
```

```python
df['date'] = df['dates'].dt.day
df['month'] = df['dates'].dt.month_name()
df['year'] = df['dates'].dt.year
df['hour'] = df['dates'].dt.hour
df['minute'] = df['dates'].dt.minute
```

```python
df = df.drop('dates' , axis=1)
df = df[df['users'] != 'group_notification']
df.head()
```

```python
total_messages = df.shape[0]
print(total_messages)
```

```python
unique_users = df.users.nunique()
print(unique_users)
```

```python
user_msg_count = df.groupby('users').size().sort_values(ascending=False)
print(user_msg_count)
```

```python
text_df = df.copy()

text_df = text_df[
    ~text_df['user_messages'].str.startswith('<', na=False)
]
text_df.head()
```

```python
from collections import Counter

words = []

for msg in text_df['user_messages']:
    words.extend(msg.lower().split())

word_freq = Counter(words)
```

```python
word_freq_df = (
    pd.DataFrame(word_freq.items(), columns=['word', 'count'])
      .sort_values(by='count', ascending=False)
)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

word_freq_df = word_freq_df[
    ~word_freq_df['word'].isin(stop_words)
]
```

```python
top_words = word_freq_df.head(20)
top_words
```
